```python
def return_stars_in_this_war(clan_tag, player_name=None, player_tag=None):
    current_war = requests.get(f"https://api.clashofclans.com/v1/clans/%23{clan_tag}/currentwar", headers={'Authorization': f"Bearer {api_key}"})
    current_war = current_war.json()
    if current_war['state'] == "inWar":
        for member in current_war['clan']['members']:
            if player_name in member['name'] or member['tag'] == player_tag:
                try:
                    attacks = member['attacks']
                    stars = 0
                    for attack in attacks:
                        stars += attack['stars']
                    return stars
                except KeyError:
                    return 0
        print("Player not found in the clan!")
    else:
        print("War has not started yet!")
        return 0
```

```python
hzuaifa_stars = return_stars_in_this_war(clan_tag, player_name="Huzaifa")
```

```python
connorwalis_stars = return_stars_in_this_war(clan_tag2, player_name="connorwallis")
print(f"Connor has earned {connorwalis_stars} stars in the war!")
```
